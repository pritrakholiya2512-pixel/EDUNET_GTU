```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import LabelEncoder

from google.colab import files
uploaded = files.upload()
df = pd.read_csv("Energy_Consumption.csv")
print("Dataset Loaded Successfully!\n")

# View Dataset
print("First 5 Rows")
print(df.head())

print("\nLast 5 Rows")
print(df.tail())

print("\nShape of Dataset")
print(df.shape)

print("\nColumn Names")
print(df.columns)

#General Information
print("\nInformation")
print(df.info())

print("\nStatistical Summary")
print(df.describe())

# General Analysis
# Dataset contains energy consumption information of different countries.

# Sno
# =Serial Number

# Country
# =Name of the country

# ISO_Code
# =Short code of country.

# Year
# =Year of recorded data.

# Total Energy Consumption (TWh)
# =Total energy used by the country.

# Renewable Energy (%)
# =Percentage of renewable energy.

# Renewable Energy Consumption (TWh)
# =Renewable energy consumed.

# Relationships between columns:
# 1.Total Energy Consumption and Renewable Energy Consumption are positively related.
# 2.Renewable Energy (%) tells how much renewable energy contributes to total energy.
# 3.Country and ISO_Code represent the same country information.
# 4.Year helps compare energy usage over time.

# Check Missing Values
print("\nMissing Values")
print(df.isnull().sum())

# Feature Engineering
df["Renewable_Efficiency"] = (df["Renewable Energy Consumption (TWh)"]
                              /df["Total Energy Consumption (TWh)"]) * 100
print("\nNew Feature Added Successfully")

# Why?
# This new column tells how efficiently a country uses renewable energy.
# Feature Engineering improves understanding of the dataset.

# Label Encoding
le = LabelEncoder()
df["Country"] = le.fit_transform(df["Country"])
df["ISO_Code"] = le.fit_transform(df["ISO_Code"])
print("\nCategorical Columns Encoded")

# Scaling Numerical Columns

scaler = StandardScaler()
numerical_columns = ["Year","Total Energy Consumption (TWh)",
                     "Renewable Energy (%)","Renewable Energy Consumption (TWh)"
                     ,"Renewable_Efficiency"]
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])
print("\nScaling Completed")

# Correlation
print("\nCorrelation Matrix")
print(df.corr(numeric_only=True))

# Data Visualization using Seaborn
sns.set_style("whitegrid")

# Histogram
plt.figure(figsize=(7,5))
sns.histplot(df["Renewable Energy (%)"], bins=20)
plt.title("Distribution of Renewable Energy Percentage")
plt.show()

#Scatter Plot
plt.figure(figsize=(7,5))
sns.scatterplot(
x=df["Total Energy Consumption (TWh)"],
y=df["Renewable Energy Consumption (TWh)"]
)
plt.title("Total vs Renewable Energy Consumption")
plt.show()

# Final Insights
#Dataset contains energy data of different countries.
#There were no missing values in this dataset.
#Feature Engineering was done by creating Renewable_Efficiency.
#Label Encoding converted Country and ISO_Code into numbers.#
#StandardScaler scaled numerical columns so that every feature has similar scale.
#Heatmap shows relationships among numerical columns.
#Scatter plot shows Total Energy Consumption and Renewable Energy Consumption are positively related.
#Histogram helps understand Renewable Energy Percentage distribution.
#Boxplot helps identify outliers.
#Dataset is now clean and ready for Machine Learning.

print("\nProject Completed Successfully!")
```
